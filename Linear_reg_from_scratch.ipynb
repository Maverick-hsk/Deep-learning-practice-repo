```python
import numpy as np
```

```python
class LinearRegression:
  def __init__(self, learning_rate=0.01, epochs = 1000):
    self.learning_rate = learning_rate
    self.epochs = epochs
    self.weight = 0
    self.bias = 0
  def predict(self, X):
    return self.weight* X + self.bias

  def compute_loss(self, y_true, y_preds):
    return np.mean((y_true - y_preds)**2)

  def fit(self,X,y):
    n = len(X)

    for i in range(self.epochs):
      y_preds = self.predict(X)
      dw = (-2/n) * np.sum(X* (y-y_preds))
      db = (-2/n) * np.sum(y-y_preds)


      self.weight -= self.learning_rate * dw
      self.bias -= self.learning_rate * db

      if i% 100 == 0:
        loss = self.compute_loss(y, y_preds)
        print(f"Epoch {i}, Loss: {loss:.4f}, Weight:{self.weight:.4f}, Bias:{self.bias:.4f}")

  def evaluate(self, X,y):
    y_preds = self.predict(X)
    loss = self.compute_loss(y, y_preds)
    print(f"Final loss: {loss:.4f}")
```

```python
# Creating some sample data
X = np.array([1,2,3,4,5], dtype= np.float32)
y = np.array([2,4,6,8,10], dtype = np.float32)

X2 = np.array([1,2,3,4,5], dtype= np.float32)
y2 = np.array([3,6,9,12,15], dtype= np.float32)

model1 = LinearRegression(learning_rate= 0.01, epochs=1000)
model2 = LinearRegression(learning_rate=0.01, epochs=1000)
```

```python
model1.fit(X,y)
model2.fit(X2,y2)
```

```python
model1.evaluate(X,y)
model2.evaluate(X2,y2)
```

```python
print("Predicted price for size  is:", model1.predict(6))
print("Predicted price for size is:", model2.predict(9))
```
